```python
import numpy as np
import pandas as pd
```

```python
DataFrame=pd.read_csv(r'time_series_ncov-Confirmed.csv')
DataFrame
```

```python
df=DataFrame.copy()
```

```python
df.describe()
```

```python
key=df.describe().keys()
key
```

```python
updated_dates=list()
for k in range(len(key)):
    count=0
    for j in range(487):
        if(df.iloc[j][k+4]==0):
            count=count+1
    per=count/487
    if(per>0.71):
        updated_dates.append(key[k])
updated_dates
```

```python
#df=pd.read_csv(r'time_series_ncov-Confirmed.csv')
for i in updated_dates:
    df = df.drop(i, axis = 1)
```

```python
df
```

```python
df_grouped = df.groupby('Country/Region').sum()
df_grouped
```

```python
region_dict = dict()
for i in range(487):
    region_dict[df.iloc[i][1]] = region_dict.get(df.iloc[i][1],0) + df.iloc[i][-1]
for region in region_dict:
    print("{} with cases:{}".format(region,region_dict[region]))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (200,100))
plt.pie(region_dict.values())
plt.title('Regionwise confirmed cases')
plt.legend(region_dict.keys(), loc='center left', prop = {'size': 64})
plt.show()
```
